# A1-LCD* simulation

First, I plotted the contact map. It show distance between residues. It can show us the topology of the protein and we can also see how the structure evolves in time.

Contact map at the start of the simulation:

````python
import MDAnalysis as mda
from my_lib import Analysis
````

````python
uni = mda.Universe("top.pdb", "sim.dcd")

first_frame = 0

protein_analysis = Analysis(uni)

protein_analysis.set_frame(first_frame)

protein_analysis.genrate_contact_map()
````

In this frame we hace a lot of side braches coming out from diagonal, this indicate the our protein is folded at initial time

Contact map at the middle of the simulation:

````python
mid_frame = int(len(uni.trajectory)/2)

protein_analysis.set_frame(mid_frame)

protein_analysis.genrate_contact_map()
````

Contact map at end of the simulation:

````python
last_frame = len(uni.trajectory) - 1

protein_analysis.set_frame(last_frame)

protein_analysis.genrate_contact_map()
````

Time evolution of the contact map:

```python
protein_analysis.set_frame(None) # it gives us the time evolution of the contact map
protein_analysis.genrate_contact_map()
```
